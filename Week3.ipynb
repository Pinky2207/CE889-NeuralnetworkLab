Formula :
ycap = f(Wt * x + y)

```python
import numpy as np
```

```python
X = np.array(
  [
   [1, 6, 8, 0],
   [33, 2, 3, 5],
   [1, 6, 6, 7],
   [2, 9, 8,4],
   [5, 6, 9, 6],
   [32, 13, 0, 2]
  ]
)

```

```python
Y = np.array(
    [
     [1, 0, 1 , 0],
     [0, 1, 0, 1]
    ]
)
```

```python
b = 0.01
```

```python
w = np.array(
    [
     [0.01, 0.01],
     [0.01, 0.01],
     [0.01, 0.01],
     [0.01, 0.01],
     [0.01, 0.01],
     [0.01, 0.01],
    ]
)

# W = np.array(
#     [
#     [0.01, 0.01, 0.01, 0.01, 0.01, 0.01],
#     [0.01, 0.01, 0.01, 0.01, 0.01, 0.01],
#     ]
# )
W_T = w.transpose() 
```

```python
print(W_T)
```

```python
def sigmoid(x):
  y = 1/(1 + np.exp(-x))
  return y 
```

```python
c = np.matmul(W_T, X) + b
print(c)
```

```python
ycap = sigmoid(c)
```

```python
print(ycap)
```

```python
loss = np.sum((ycap - Y), axis=1)
```

```python
print(loss)
```
