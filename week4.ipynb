```python
def dldw(X, ycap, Y, k):
  backwardprop = np.matmul(X, (ycap - Y).transpose())
  return(backwardprop/k)
```

```python
def dldb(Y,ycap,k):
  biasweight = np.sum((ycap-Y), axis=1)
  return(biasweight/k)
```

```python
def updatingWeight(W,learning_rate, backwardprop):
  weight = W - (learning_rate * backwardprop)
  return(weight)
```

```python
def updatingBias(b, learning_rate, biasweight):
  bias = b - (learning_rate * biasweight)
  return(bias)
```
